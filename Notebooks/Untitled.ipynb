```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tmp = pd.read_csv('../Patent Demo (New)/Data/subset_post_2018.csv')
```

```python
tmp.index = pd.DatetimeIndex(tmp.date)
```

```python
tmp.date.groupby(pd.Grouper(freq='2m')).count().plot()
plt.show()
```

```python
tmp_df = tmp[['network' in x for x in tmp.abstract]]
max_values = 8_000
len_df = len(tmp_df)
tmp_df = tmp_df.iloc[::len_df//max_values]
```

```python
tmp_df = tmp_df.groupby([pd.Grouper(freq='2m'),'organization']).count().id.sort_values(ascending=False).unstack(fill_value=0).stack()
x_axis = tmp_df.index.get_level_values(0).unique()
print(x_axis)
groups = tmp_df.index.get_level_values(1).unique()
traces = []
```
