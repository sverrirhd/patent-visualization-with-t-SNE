```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Sækja öll dataset


```python
# ATH: Rúmlega 5GB af data sem þarf að loada
dirName = '../../Data/Research Dataset/All Data/'
fileNames = ['assignee',
             'assignment',
             'assignment_conveyance',
             'assignor',
             'documentid',
             'documentid_admin']

df_dict = {}

for fileName in fileNames:
    print('Loading %s file'%(fileName))
    fileNameString = dirName+fileName+'.csv'
    print('directory: %s'%(fileNameString))
    tmp_df = pd.read_csv(fileNameString)
    tmp_df.index = tmp_df.rf_id
    df_dict[fileName] = tmp_df
    

```

# 2. Henda öllu sem er ekki norðurlöndin

```python
nordicCountries = ['SWEDEN','FINLAND','NORWAY','DENMARK','ICELAND']
nordic_patent_rf_ids = df_dict['assignee'][[x in nordicCountries for x in  df_dict['assignee'].ee_country]].rf_id
print('Number of nordic patents on file:', len(nordic_patent_rf_ids))
df_dict_nordic = {}
for df_name in df_dict:
    df_dict_nordic[df_name] = df_dict[df_name].loc[nordic_patent_rf_ids]
```

# 3. Sameina dataset
    

```python
df_nordic = pd.DataFrame(index=nordic_patent_rf_ids)
for df_name in ['assignee','assignment']:
    print(df_dict_nordic[df_name].columns)
    df_nordic = pd.merge(df_nordic,df_dict_nordic[df_name],how='left',on='rf_id')
```

### Getum séð að lang flest patent númer koma aðeins einu sinni fyrir en einhver koma mjög oft fyrir

Það eru 50000 raðir í gagnagrunninum sem hafa sama RF_ID


```python


non_unique_rf_ids = sum(df_nordic.rf_id.value_counts().value_counts().index[1:]*df_nordic.rf_id.value_counts().value_counts()[1:].values)
print('Number of rows with non-unique rf_id:',non_unique_rf_ids)
print('Number of rf_ids with single row in dataframe:',df_nordic.rf_id.value_counts().value_counts().iloc[0])

print('Most used rf_ids and their count:')
print(df_nordic.rf_id.value_counts()[:5])
print()
```

```python
most_common_rf_id = df_nordic.loc[df_nordic.rf_id.value_counts().index[0]]
print('Number of patents with most common rf_id:',len(most_common_rf_id))
print('Few rows of most common rf_id:')
print('Number of unique values for each column')
for col in most_common_rf_id.columns:
    print(len(most_common_rf_id[col].unique()),'\t',col)

print('Number of rows with a complete duplicate row',sum([x == True for x in most_common_rf_id.duplicated()]))
print('Number of rows with no complete duplicate row',sum([x == False for x in most_common_rf_id.duplicated()]))
```
